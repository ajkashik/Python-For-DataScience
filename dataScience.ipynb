```python
#combining dataset:: concate and append

import pandas as pd
import numpy as np
def ramadan(col,ind):
    x={c:[str(c)+str(i) for i in ind] for c in col}
    return pd.DataFrame(x,ind)
ramadan('WXYZ',range(4))
```

```python
def love(cols,ind):
    v={p:[str(p)+str(e) for e in ind] for p in cols}
    return pd.DataFrame(v,ind)
love("MNOP",range(3))
```

```python
def ind_col(col,ind):
    data={d:[str(d)+str(e) for e in ind] for d in col}
    return pd.DataFrame(data,ind)
ind_col("PQRS",range(1,4))
```

```python
def fuck(col,ind):
    data={x:[str(x)+str(i) for i in ind] for x in col}
    return pd.DataFrame(data,ind)
fuck("DMPK",range(2))
```

```python
# multiple legend adding
fig,ax=plt.subplots()
lines=[]
styles=['x','v','d','y']
x=np.linspace(0,10,1000)

for i in range(4):
    lines=lines+ax.plot(x,np.sin(x-i),styles[i],color='red')
ax.axis('equal')
ax.legend(lines[:2],['line A','line B'],loc='upper left')
from matplotlib.legend import Legend
leg=Legend(ax,lines[:2],['line C','line D'],loc='lower right')
ax.add_artist(leg);
```

```python
import pandas as pd
from sklearn.datasets import load_iris
iris=load_iris()
```

```python
df=pd.DataFrame(iris.data,columns=iris.feature_names)
df.head()
```

```python
df['target']=iris.target
df.head()
```

```python
iris.target_names
```

```python
df[df.target==1].head()
```

```python
df[df.target==2].head()
```

```python
df['flower_name']=df.target.apply(lambda x: iris.target_names[x])
df.head()
```

```python
df0=df[df.target==0]
df1=df[df.target==1]
df2=df[df.target==2]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X=df.drop(['target','flower_name'],axis='columns')
X.head()
```

```python
y=df.target
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
from sklearn.svm import SVC
model=SVC(gamma=10)
```
